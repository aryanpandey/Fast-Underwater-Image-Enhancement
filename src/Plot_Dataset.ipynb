```python
import torch
from PIL import Image
from dataset import EUVP
import matplotlib.pyplot as plt
```

```python
def plot_images(images, targets, normalize = False):

    n_images = len(images)
    
    
    fig, axs = plt.subplots(n_images, 2, figsize = (10, 12), sharex = True, sharey = True)
    
    for i in range(n_images):
        image = images[i]
        target = targets[i]

        if normalize:
            image_min = image.min()
            image_max = image.max()
            image.clamp_(min = image_min, max = image_max)
            image.add_(-image_min).div_(image_max - image_min + 1e-5)

        axs[i, 0].imshow(image)
        axs[i, 1].imshow(target)
```

```python
data = EUVP(root_dir = 'Data/EUVP_Dataset/Paired', img_dim = 256, transform = False)
```

```python
N_IMAGES = 3

images, targets = zip(*[(image, target) for image, target in 
                           [data[i] for i in range(N_IMAGES)]])

plot_images(images, targets)
```
